OpenBLAS Thread Tuning
======================

OpenBLAS threading thresholds are somewhat opaque, and we don't have much evidence as to what are genuinely useful values for many of the algorithms.  This notebook aims to explore the parameter space to find optimal thresholds for various problems.  To start, we will need a patched version of OpenBLAS that creates and gives public access to internal thresholds, which you can find [here][1]

[1]: https://github.com/staticfloat/OpenBLAS/tree/sf/threadingparams

```julia
# This is a reflection of what is defined in common_thread.h
@struct type openblas_threading_params
    num_threads::Uint32
    threshold_surpassed::Uint32
    gemm_threshold::Uint32
end

# Interrogate OpenBLAS about its threading parameters
function get_threading_params()
    param_ptr = ccall((:openblas_get_threading_params, Base.libblas_name), Ptr{Void}, ())
    
    # Load num_threads and gemm_threshold
    num_threads = unsafe_load(convert(Ptr{Cint}, param_ptr), 1)
    threshold_surpassed = unsafe_load(convert(Ptr{Cint}, param_ptr), 2)
    gemm_threshold = unsafe_load(convert(Ptr{Cint}, param_ptr), 3)
    
    return openblas_threading_params(num_threads, threshold_surpassed, gemm_threshold)
end

function set_threading_params(p::openblas_threading_params)
    iostr = IOBuffer()
    pack(iostr, p)
    
    ccall((:openblas_set_threading_params, Base.libblas_name), Void, (Ptr{Void},), iostr.data)
end

# Use this to figure out how many (if any) BLAS calls we just made were multithreaded
macro mt_count(ex)
    quote
        local pre_count = get_threading_params().threshold_surpassed
        $(esc(ex))
        local post_count = get_threading_params().threshold_surpassed
        post_count - pre_count
    end
end

# Get params
params = get_threading_params()
@show params

# tweak them a litle
params.gemm_threshold = 20
params.num_threads = 4
set_threading_params(params)

# Show that they've actually changed!
params = get_threading_params()
@show params;
```

```julia
# Now let's see what gemm_threshold actually does!

NUM_TRIALS=100
prexp_sizes = linspace(1,9,14)
sizes = int(round(2.^prexp_sizes))
times = Array(Float64, length(sizes), length(sizes), NUM_TRIALS)
mt_track = Array(Int64, length(sizes), length(sizes))

# If I'm timing something for Tim, I'm sure this macro name doesn't add confusion at all
macro tim(ex)
    quote
        local t0 = time_ns()
        local g0 = Base.gc_time_ns()
        $(esc(ex))
        local t1 = time_ns()
        local g1 = Base.gc_time_ns()
        
        # Let's ignore gc time for now
        (t1-t0 - (g1 - g0))/1e9
    end
end

function test_gemm(n, k, n_idx, k_idx)
    A = randn(n, k)
    B = randn(k, n)
    C = Array(Float64, n, n)

    BLAS.gemm!('N','N',1.0,A,B,0.0,C)
    
    # Simultaneously warm up and check if this operation is going to be multithreaded
    mt = @mt_count BLAS.gemm!('N','N',1.0,A,B,0.0,C)
    mt_track[n_idx, k_idx] = mt

    for i = 1:NUM_TRIALS
        times[n_idx, k_idx, i] = @tim BLAS.gemm!('N','N',1.0,A,B,0.0,C)
    end
end

for n_idx in 1:length(sizes)
    for k_idx in 1:length(sizes)
        #test_gemm(sizes[n_idx], sizes[k_idx], n_idx, k_idx)
        @printf "[%dx%d]:   \t %.3e\n" sizes[n_idx] sizes[k_idx] times[n_idx, k_idx]
    end
end

@printf "done\n"
```

```julia
x = 1:length(sizes)
xgrid = repmat(x,length(sizes))
ygrid = transpose(repmat(x,1,length(sizes)))[:]

c = log(minimum(times,3))
c = mt_track
p1 = plot(x=xgrid, y=ygrid, color=c, Geom.rectbin)

display(p1);
```
