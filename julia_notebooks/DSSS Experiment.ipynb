```julia
function load_grc_sink(path::String, datatype::DataType)
    fd = open(path, "r")
    datalen = filesize(path)
    
    return mmap_array(datatype, (ifloor(datalen/sizeof(datatype)),), fd );
end

data = load_grc_sink("/Users/sabae/src/wisee2d/80211/data/beacon_80211g.complex64", Complex64)
preamble = data[3000:4000]

plot(abs(preamble))
title("preamble temporal response");
```
