```julia
# let's start with a random seed.  Yes, this is cheating since I'm using rand() :P
function xorshift(N, s = rand(Uint64, 16))
    x = Array(Uint64, N)
    p = 0
    for i = 1:N
        # One-based indexing strikes again
        s0 = s[p + 1]
        p = (p + 1)%16
        s1 = s[p + 1]

        # Make sure you use $ instead of ^ here!
        s1 $= s1 << 31
        s1 $= s1 >> 11
        s0 $= s0 >> 30
        s[p + 1] = s0 $ s1

        # I love magic constants
        x[i] = s[p + 1] * 1181783497276652981
    end
    return x
end
```

```julia
using PyPlot

xor_mean = Float64[]
xor_min = Float64[]
xor_max = Float64[]

rand_mean = Float64[]
rand_min = Float64[]
rand_max = Float64[]

num_trials = 10

for N in int64(logspace(2, 9, 10))
    println("Testing generation of $N 64-bit numbers")
    xor_trial_times = Float64[]
    for n in 1:num_trials
        start = time()
        xorshift(N)/2.0^64
        push!(xor_trial_times, time() - start)
    end
    push!(xor_mean, mean(xor_trial_times))
    push!(xor_min, minimum(xor_trial_times))
    push!(xor_max, maximum(xor_trial_times))
    
    rand_trial_times = Float64[]
    for n in 1:num_trials
        start = time()
        rand(Uint64,N)/2.0^64
        push!(rand_trial_times, time() - start)
    end
    push!(rand_mean, mean(rand_trial_times))
    push!(rand_min, minimum(rand_trial_times))
    push!(rand_max, maximum(rand_trial_times))
end

figure()
plot(xor_mean, "g")
plot(xor_min, "g--")
plot(xor_max, "g--")

plot(rand_mean, "b")
plot(rand_min, "b--")
plot(rand_max, "b--")

legend(["xor", "xor_min", "xor_max", "rand", "rand_min", "rand_max"])
```
